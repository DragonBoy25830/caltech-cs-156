```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
```

```python
x = np.array([[1, 0], [0, 1], [0, -1], [-1, 0], [0, 2], [0, -2], [-2, 0]])
y = [-1, -1, -1, 1, 1, 1, 1]
c = ['blue' if y < 0 else 'magenta' for y in y]

plt.scatter(x[:, 0], x[:, 1], c=c)
```

```python
z = np.array([(x2 * x2 - 2 * x1 - 1, x1 * x1 - 2 * x2 + 1) for (x1, x2) in x])
c = ['blue' if y < 0 else 'magenta' for y in y]

plt.scatter(z[:, 0], z[:, 1], c=c)
plt.scatter([-2, -2], [-2, 3])
```

```python
def run_svm(C, Q):
    model = SVC(C=C, kernel='poly', degree=Q, gamma=1)
    model.fit(x, y)
    
    return np.sum(model.n_support_)
```

```python
possible_c_values = [10 ** i for i in range(-10, 11)]

for C in possible_c_values:
    num_support_vectors = run_svm(C, 2)
    print(f"C: {C} --> {num_support_vectors} support vectors")
```
