```python
import random
import statistics as stats
import matplotlib.pyplot as plt
```

```python
num_flips = 10
num_coins = 1000
num_trials = 100000
```

```python
def flip_coin(num_flips):
    flips = []
    for _ in range(num_flips):
        flip = random.choice([0, 1])
        
        if flip == 0:
            flips.append("H")
        else:
            flips.append("T")
    return flips
```

```python
v1 = []
vrand = []
vmin = []
```

```python
def find_min_coin(coin_flips):
    min_freq = 100
    min_index = 0
    
    for i in range(len(coin_flips)):
        coin_flip = coin_flips[i]
        
        if coin_flip.count("H") < min_freq:
            min_index = i
            min_freq = coin_flip.count("H")
    
    return coin_flips[min_index]
```

```python
def run_experiment():
    total_coins_flips = []
    
    for _ in range(num_coins):
        total_coins_flips.append(flip_coin(num_flips))
    
    first_coin = total_coins_flips[0]
    random_coin = random.choice(total_coins_flips)
    min_coin = find_min_coin(total_coins_flips)
    
    v1.append(first_coin.count("H") / num_flips)
    vrand.append(random_coin.count("H") / num_flips)
    vmin.append(min_coin.count("H") / num_flips)
```

```python
for _ in range(num_trials):
    run_experiment()
```

```python
stats.mean(vmin)
```
