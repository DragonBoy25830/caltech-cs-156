```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_points = 1000
delta = 1e-5
```

Random Line generated as target function

```python
ax = np.random.uniform(-1, 1 + 1e-5)
bx = np.random.uniform(-1, 1 + 1e-5)
ay = np.random.uniform(-1, 1 + 1e-5)
by = np.random.uniform(-1, 1 + 1e-5)

f = np.array([bx - ax, by - ay])
```

```python
def target_function(x, y):
    cross_product = (bx - ax)*(y - ay) - (by - ay)*(x - ax)
    
    if cross_product > 0:
        return -1
    else:
        return 1

ax
```

```python
def sign_inner_product(w, x):
    return np.sign(np.dot(w, x))
```

```python
x1 = np.random.uniform(-1, 1 + delta, num_points)
x2 = np.random.uniform(-1, 1 + delta, num_points)

xn = np.column_stack((x1, x2))
yn = [target_function(x, y) for (x, y) in xn]

plt.plot([i[0] for i in xn], [i[1] for i in xn], 'o')
plt.axline((ax, bx), (ay, by))
```

```python
weights = np.zeros(2)
```

```python
misclassified_points = list(zip(xn, yn))
random_index = np.random.randint(0, num_points)
misclassified_point = (xn[random_index], yn[random_index])
misclassified_point
```

```python
iterations = [weights]
```

```python
def PLA(xn, yn):
    random_index = np.random.randint(0, num_points)
    misclassified_point = (xn[random_index], yn[random_index])
    
    while sign_inner_product(iterations[-1], misclassified_point[0]) == np.sign(misclassified_point[1]):
        random_index = np.random.randint(0, num_points)
        misclassified_point = (xn[random_index], yn[random_index])
        
    w = iterations[-1] + misclassified_point[1] * misclassified_point[0]
    iterations.append(w)
```

```python
for i in range(100000):
    PLA(xn, yn)
```
